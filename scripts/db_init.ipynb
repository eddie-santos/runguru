```python
import pymysql as mdb

con = mdb.connect('localhost', 'root', '', 'stravadb') #host, user, password, #database

with con:
    cur = con.cursor()
    cur.execute("DROP TABLE IF EXISTS Strava")
    cur.execute("CREATE TABLE Strava(Id INT PRIMARY KEY AUTO_INCREMENT,user INT,name VARCHAR(25),age INT,activity INT,time INT,pace INT)")
#     cur.execute("INSERT INTO Strava(name,age,user,activity,time,pace) VALUES('Jack London','25','34234','37473335','3:30:00','6:57')")

with con:
    cur = con.cursor()
    cur.execute("SELECT * FROM Strava")
    rows = cur.fetchall()
    for row in rows:
        print row
```
